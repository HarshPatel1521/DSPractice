```python
values = [['Rohan', 455], ['Elvish', 250], ['Deepak', 495],
         ['Sai', 400], ['Radha', 350], ['Vansh', 450]]

dval = pd.DataFrame(values, columns = ['Name', 'Univ. Marks'])

dval = dval.assign(Percentage = lambda x: (x['Univ. Marks'] / 500 * 100))

dval.sort_values(by = ['Percentage'])
```
